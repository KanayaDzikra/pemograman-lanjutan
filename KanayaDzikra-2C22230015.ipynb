Problem (1) Kelipatan 10 terkecil

```python
# Input bilangan N
N = int(input("Masukkan N: "))

# Mencari kelipatan 10 terkecil yang lebih besar dari N
i = N + 1
while i % 10 != 0:  # selama i belum menjadi kelipatan 10
    i += 1  # Tambah 1

# Output hasil
print("output:", i)
```

- Interpretasi coding:

	‚Ä¢	Kita mulai dengan kelipatan awal 10.

  ‚Ä¢	Menggunakan loop while, kita akan mengalikan angka tersebut dengan 10 sampai nilainya lebih besar dari N.

- Interpretasi Hasil:

  ‚Ä¢	jika N=5, hasilnya akan 10 (karena 10 adalah kelipatan terkecil dari 10 yang lebih besar dari 5).

  ‚Ä¢	Jika N=1234, maka program akan melanjutkan perhitungan hingga mencapai 10000 (kelipatan 10 terkecil lebih besar dari 1234).


Untuk soal ini, loop while lebih baik digunakan karena kita tidak tahu secara pasti berapa kali kita harus mengalikan dengan 10 hingga mencapai angka yang lebih besar

Problem (2) Segitiga Angka

```python
N=int(input("Masukkan bilangan N: "))
for i in range(1,N+1,1):
  for j in range(1,i+1):
    print(j,end="")
  print()
```

- Interpretasi coding:

- Loop luar (for i in range(1, N + 1)) mengatur jumlah baris. Setiap kali i bertambah, kita akan mencetak satu baris baru.
- Loop dalam (for j in range(1, i + 1)) mengatur angka yang dicetak pada setiap baris. Pada baris ke-N, kita mencetak angka dari 1 hingga N.
- Fungsi print(j, end=" ") memastikan angka-angka dicetak dalam satu baris tanpa pindah baris, sedangkan print() setelah loop dalam digunakan untuk berpindah ke baris baru setelah satu baris angka selesai dicetak
- Interpretasi Hasil:

Untuk soal ini, loop bersarang (nested loop) adalah pilihan terbaik karena kita perlu mencetak sejumlah angka yang bervariasi di setiap baris (angka 1 hingga N pada baris ke-N).

Perbandingan Loop pada Kedua Soal:

	‚Ä¢	Kelipatan 10 Terkecil: Loop yang lebih cocok digunakan adalah loop while, karena kita tidak tahu berapa kali kita harus mengalikan dengan 10 hingga mencapai angka yang lebih besar dari Ôøº. Loop ini lebih dinamis dan kondisional.
	‚Ä¢	Segitiga Angka: Nested loop for lebih cocok digunakan karena kita mengetahui jumlah iterasi yang pasti (berdasarkan nilai Ôøº), dan tiap iterasi memiliki batas yang jelas (jumlah angka pada setiap baris).

Kesimpulan Karakteristik Loop:

	‚Ä¢	While loop cocok untuk kasus ketika jumlah iterasi tidak diketahui sebelumnya dan bergantung pada kondisi yang terpenuhi.
	‚Ä¢	For loop lebih cocok ketika jumlah iterasi sudah diketahui atau mudah diprediksi dari input.

```python
# Input nilai N, A, B, dan C
N = int(input("Masukkan nilai N: "))
A = int(input("Masukkan nilai A: "))
B = int(input("Masukkan nilai B: "))
C = int(input("Masukkan nilai C: "))

# Loop dari 1 sampai N
for x in range(1, N + 1):
    output = ""

    # Cek kelipatan A, B, dan C
    if x % A == 0:
        output += "Siap"
    if x % B == 0:
        output += "Bang"
    if x % C == 0:
        output += "Jago"

    # Jika tidak ada yang diprint (bukan kelipatan A, B, atau C), cetak angka
    if output == "":
        output = str(x)

    # Print hasil untuk x
    print(output, end=" ")
```
